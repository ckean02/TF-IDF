```python
TEI_NS = "http://www.tei-c.org/ns/1.0"
XML_NS = "http://www.w3.org/XML/1998/namespace"

NAMESPACES = {
    "tei": TEI_NS,
    "xml": XML_NS,
}
```

```python
for file in files:
    print(file)
    tree = etree.parse(file)
    text = tree.xpath(f"//tei:div[@subtype='card']//text()", namespaces=NAMESPACES)
    
    cleaned_text = []
    for t in text:
        if t.strip() != "":
            cleaned_text.append(t)

    if len(cleaned_text) > 0:
        with open(str(file).split("/")[-1].replace(".xml", ".txt"), "w+") as f:
            f.write('\n'.join(cleaned_text))
    
```

TF shows how frequently a term appears in a document relative to its length. This tells us how important a term is within that specific document. IDG measures how rare or common a term is across all documents. More common terms = lower IDF values. The TF-IDF score (TF × IDF) balances these two factors: High TF-IDF = Terms that appear frequently in a specific document but rarely in other documents. Low TF-IDF = Terms that appear rarely in a document or commonly across all documents. For example, character names or themes unique to either the Iliad or Odyssey would have high TF-IDF scores in their respective texts. Rather than just counting raw occurrences, TF-IDF helps distinguish between truly significant terms and nondistinctive common words.
